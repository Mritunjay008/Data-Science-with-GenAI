### **Understanding Line Plot with Confidence Interval in Seaborn**
In **`sns.relplot(x='timepoint', y='signal', data=fmri, kind='line')`**, the **line** represents the **mean trend** of the `signal` over `timepoint`, and the **shaded region** represents the **confidence interval (CI)** around the mean.  

### **Understanding the Graph:**  
- **Line** → Shows the average value of `signal` at each `timepoint`.  
- **Shaded Region** → Represents the **uncertainty** or **variability** (default: 95% confidence interval) around the mean.  
- **Why is it useful?** It helps visualize the **trend** and **variation** in data over time.  

If you want to remove the shaded region, you can set `ci=None` in the function. 🚀

### **Different Types of `ci` (Confidence Interval) in Seaborn**  

The **`ci` (confidence interval)** parameter in Seaborn controls the shaded region around the line in `sns.lineplot()` or `sns.relplot(kind='line')`. Different options include:  

- **`ci="sd"`** → Shows **standard deviation**, indicating data spread.  
- **`ci=95`** (default) → Plots a **95% confidence interval**, meaning there is a 95% probability that the true mean lies within the shaded region.  
- **`ci=None`** → Removes the confidence interval, displaying only the line.  
- **`ci=50`** → Plots a **50% confidence interval**, making the shaded region narrower.  

Choosing `ci` depends on whether you want to **highlight variability (`sd`)** or **statistical confidence (`95`, `50`)** in your data. 🚀

````python
sns.relplot(x='timepoint', y='signal', data=fmri, kind='line')
````

- Used for **visualizing categorical data**.  
- Supports multiple plot types via `kind` (e.g., **bar, box, violin, strip, swarm**).  
- **X-axis → Categorical variable**, **Y-axis → Numerical variable**.

### **`sns.jointplot()` – Relationship & Distribution Between Two Variables**  

`jointplot()` is used to visualize the **relationship between two numerical variables** along with their **individual distributions**. It combines a **scatter plot** (for relationships) with **histograms or KDE plots** (for distributions).  

### **Key Features:**  
- Default plot: **scatter plot + histograms**.  
- Use **`kind="kde"`** for density estimation.  
- Use **`kind="hex"`** for hexbin plots (good for large datasets).  
- Marginal plots on x and y axes show individual distributions.  

### **Example Usage:**  
```python
sns.jointplot(x='height', y='weight', data=df, kind='scatter', hue='gender')
```
This will plot a scatter plot of **height vs. weight** with **separate distributions** for each variable. 🚀

````python
# we want to understand distribution of 2 data
# column >> univariate(1 column), bivariate(2 column)

sns.jointplot(x=tips.total_bill, y = tips.tip)
# It draw a plot of two variables with distribution  >> bivariate plots
````

````python
sns.jointplot(x=tips.total_bill, y = tips.tip, kind='reg')
# reg >> regression line  to understand relationship
````

````python
sns.jointplot(x=tips.total_bill, y = tips.tip, kind='hex')
````

````python
sns.jointplot(x=tips.total_bill, y = tips.tip, kind='kde')
# kde >> kernal density estimation

# The plot is contour plot >> used in metrological plot
````

### **Correlation**  

Correlation measures the **strength and direction** of the relationship between two numerical variables. A **positive correlation** means both increase together (e.g., height & weight), while a **negative correlation** means one increases as the other decreases (e.g., temperature & sweater sales). If there's **no correlation**, the variables are unrelated. It is commonly measured using **Pearson’s correlation** for linear relationships. Example: `df.corr()` in pandas shows correlation between all numerical columns. 🚀

````python
df1.corr()
````

### **Heatmap**  

A **heatmap** is a graphical representation of data where values are shown using color intensity. It is often used to visualize **correlation matrices**, highlighting relationships between numerical variables. Darker or brighter colors indicate stronger correlations.  

**Example:**  
A heatmap of `df.corr()` helps identify highly correlated features in a dataset for feature selection in machine learning. 🚀

````python
# Heatmap >> value encoded with diffrent colors

sns.heatmap(df1.corr(), cmap='coolwarm', annot=True)
````
