## Scatter Plot
A **scatter plot** is used to visualize the relationship between two numerical variables by plotting data points on a 2D graph. It helps identify patterns, correlations (positive, negative, or none), and trends between variables. For example, in analyzing sales vs. advertising spend, a scatter plot can reveal whether increased spending leads to higher sales. It is also useful for detecting outliers, where certain points deviate significantly from the trend. Scatter plots are widely used in data analysis, statistics, and machine learning to understand dependencies and make predictions based on observed data distributions.

```python
plt.scatter(x, y)
```

## Line Plot
A **line plot** is used to show trends or changes over time by connecting data points with a continuous line. It is ideal for visualizing time series data, such as stock prices, temperature changes, or website traffic over days, months, or years. Unlike scatter plots, which show individual data points, line plots emphasize the overall trend, making it easier to identify patterns like upward or downward movement, seasonal fluctuations, or cyclic behavior. They are widely used in financial analysis, forecasting, and scientific research where tracking progression over time is essential.

### Histogram >>
A **histogram** is a powerful tool for visualizing the distribution of continuous numerical data, helping to uncover patterns, trends, and outliers. By grouping data into bins and representing frequencies with bars, it provides insights into data spread, symmetry, skewness, and variations. Whether identifying peaks, detecting gaps, or comparing datasets, histograms make it easy to analyze and interpret large amounts of information at a glance.
</br>
</br>
**Continuous numerical data** consists of measurable values that can take any number within a range, including decimals. Examples: height, temperature, time, and weight.
</br>
</br>
**Discrete numerical data** consists of countable values that have distinct, separate numbers (no decimals). Examples: number of students in a class, goals scored in a match, or cars in a parking lot.

```python
data = np.random.randn(1000)
plt.hist(data)
plt.show()
# Histogram will divide data in 10 intervals/bins generally
# In which interval, how many data points are there
```

```python
data = np.random.randn(1000)
plt.hist(data, color='b', bins=20)
plt.show()
```

```python
# Stacked histogram >> 2 histograms together
data1 = np.random.normal(0, 1, 5000)
data2 = np.random.normal(3, 2, 5000)
```

```python
plt.hist([data1, data2], bins = 20, color=['b', 'g'], label=['dist1', 'dist2'])
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.title("Stacked histogram")
plt.show()
```

### Bar Plot >> 
A **bar plot** represents categorical data using rectangular bars, where the height (or length) of each bar corresponds to the value or frequency of a category. It is used to compare different groups, identify trends, and visualize distributions in a clear and simple way. Bar plots can be **vertical** or **horizontal** and are widely used in business, statistics, and data analysis.

```python
# Bar plot >> To see the distribution of categorical variables
x = ["a", "b", "c", "d"] # Name of students
y = [4, 6, 1, 7] # Present in class in a month

plt.bar(x, y, color='g')
plt.grid()
plt.show()
```

```python
# Horizontal Bar plot
x = ["a", "b", "c", "d"] # Name of students
y = [4, 6, 1, 7] # Present in class in a month

plt.barh(x, y, color='g')
plt.grid()
plt.show()
```

### Pie-Chart >>
A **pie chart** is a circular chart divided into slices, where each slice represents a proportion of a whole. It is used to **visualize percentage-based data** and compare different categories. The size of each slice is proportional to its value, making it easy to see dominant categories at a glance. Pie charts are commonly used in business, surveys, and data analysis to show distributions clearly.

```python
# Pi-Chart >> to understand the percent distribution of data

langs=["c", "python", "java"]
students = [20, 100, 40]

plt.pie(students, labels=langs, autopct="%1.1f%%")
plt.title("Pi-Chart")
plt.show()
```

```python
# what if we want to focus on 1 segment
explode = (0.0, 0.1, 0.1)
langs=["c", "python", "java"]
students = [20, 100, 40]

plt.pie(students, labels=langs, autopct="%1.1f%%", explode=explode)
plt.title("Pi-Chart")
plt.show()
```

```python
# To make it more beautiful implement shadow
explode = (0.0, 0.1, 0.1)
langs=["c", "python", "java"]
students = [20, 100, 40]

plt.pie(students, labels=langs, autopct="%1.1f%%", explode=explode, shadow=True)
plt.title("Pi-Chart")
plt.show()
```

### Box Plot
A **box plot** (or **box-and-whisker plot**) is a graphical representation of data distribution based on five summary statistics: **minimum, first quartile (Q1), median (Q2), third quartile (Q3), and maximum**. It helps identify **central tendency, variability, outliers, and skewness** in a dataset. Box plots are widely used in statistics and data analysis to compare multiple distributions efficiently.

```python
# boxplot >> to check the distribution of data and outliers(extreme/abnormal) values
df.boxplot(column='Balance')
```

### Violin Plot >>
A **violin plot** is a combination of a **box plot** and a **density plot** that shows the distribution of numerical data. It displays the **median, quartiles, and density estimation**, making it useful for understanding both the spread and shape of the data. Violin plots are especially helpful when comparing multiple datasets, as they reveal patterns, skewness, and multimodal distributions more effectively than box plots.

```python
plt.violinplot(df.Balance)
```

### Area Plot 
An **area plot** is similar to a line plot but with the area beneath the line filled with color. It is used to visualize **trends over time**, showing how values accumulate or change. Area plots are useful for comparing multiple datasets and highlighting the magnitude of differences. They are commonly used in financial data, population growth analysis, and time-series data visualization.

```python
# Area Plot >> Variables changed over time

years = [2021, 2018, 2022, 2023, 2017]
catg1 = [20, 45, 34, 76, 34]
catg2 = [50, 45, 42, 66, 54]
catg3 = [67, 67, 34, 28, 56]

plt.stackplot(years, catg1, catg2, catg3, labels=["Category 1", "Category 2", "Category 3"], colors=['Magenta', 'black', "lightgreen"])
plt.xlabel("Years")
plt.xlabel("Product selled")
plt.title("Area Plot")
plt.legend()
plt.show()
```

### **How to Decide Which Type of Plot to Use**  

- **Comparison:**  
  - **Bar Plot** â†’ Compare categorical data  
  - **Line Plot** â†’ Show trends over time  

- **Distribution:**  
  - **Histogram** â†’ Visualize the frequency of numerical data  
  - **Box Plot** â†’ Show summary statistics and outliers  
  - **Violin Plot** â†’ Display data distribution with density  

- **Relationship:**  
  - **Scatter Plot** â†’ Show correlation between two numerical variables  
  - **Relplot** â†’ Analyze relationships flexibly  

- **Composition:**  
  - **Pie Chart** â†’ Represent proportions of a whole  
  - **Stacked Area Plot** â†’ Show how values change over time
